## Funciones

### Las funciones son bloques de código reutilizables que realizan una tarea específica. Permiten organizar y modularizar el código, haciéndolo más legible y fácil de mantener. Al utilizar funciones, se pueden evitar repeticiones y hacer que el código sea más eficiente.

### Como se forman las funciones?

#### 1. Nombre de la funcion
#### 2. Parámetros (o variables de entrada)
#### 3. Estructura de la funcion (condicionales, bucles, etc)
#### 4. Variables locales (solo funcionan dentro de la funcion)
#### 5. Valor de retorno (return)


```python
## Ejemplo

#1. Suma

def suma(a, b):  # Encabezado de la función con nombre "suma" y parámetros "a" y "b"
    resultado = a + b  # Cuerpo de la función y variable local "resultado"

    return resultado # Valor de retorno

# Llamada a la función
total = suma(5, 3)
#print(total)
total_2 = suma(8,6)
#print(total_2)

```

```python
print(total)
```

```python
#2. Suma sin return

def suma(a, b):  # Encabezado de la función con nombre "suma" y parámetros "a" y "b"
    resultado = a + b  # Cuerpo de la función y variable local "resultado"
     # Valor de retorno

# Llamada a la función
total = suma(5, 3)
print(total)
total_2 = suma(8,6)
print(total_2)
```

```python
#3. Suma sin return pero con print

def suma(a, b):  # Encabezado de la función con nombre "suma" y parámetros "a" y "b"
    resultado = a + b  # Cuerpo de la función y variable local "resultado"
     # Valor de retorno
    print(resultado)

# Llamada a la función
total = suma(5, 3)
print(total)
total_2 = suma(8,6)
print(total_2)

#### Es necesaria la inclusion del return para que la variable a la que se desea asignar el valor pueda seguir siendo utilizada
```

```python
#4. Funcion con texto

def name(string1, string2):
  full_string = string1 +  " " + string2
  return full_string

nombre = name("Andres","Salas")
nombre_2 = name("Carlos","Solano")
print(nombre_2)
```

```python
#4. Funcion que devuelva 2 o mas valores

def getPerson(name, age, country):
    return name,age,country

name,age,country = getPerson("Pedro",25,"Costa Rica")
```

```python
#5. Operaciones basicas

def operaciones_basicas(a, b):
    suma = a + b
    resta = a - b
    multiplicacion = a * b
    if b != 0:
        division = a // b
    else:
        division = "Indefinido"
    return suma, resta, multiplicacion, division

## Llamar la funcion con los valores
resultados = operaciones_basicas(10, 0)
print(resultados)
print(type(resultados))

## Asignar cada resultado a una variable
s, r, m, d = operaciones_basicas(10, 0)
print(s)
print(r)
print(m)
print(d)
```

## Funciones lambda

#### Una función lambda es una pequeña función anónima, es decir, es una función sin nombre. A diferencia de una función definida con def, la función lambda puede tener cualquier número de argumentos, pero solo puede tener una expresión. No puede contener comandos, y devuelve el valor de esa expresión por defecto.

```python
# Estructura

lambda argumentos: expresion
```

```python
# Ejemplo 1

suma = lambda a, b: a + b ## Variable = lambda (arg1) (arg2): a + b
print(suma(4, 2))
print(suma(5,6))
```

```python
resta = lambda c,d: c-d ## Resta
print(resta(4,2))
```

```python
# Ejemplo 2

greet = lambda : print('Hello World')

# call lambda function
greet()
```

```python
# Ejemplo 3

# lambda accepta un valor llamado 'name'

user = lambda name : print('Hola,', name, '!')

# lambda call
nombre = user('Carlos')
user('Pedro')
print(nombre)
```

```python
## Ejemplo 4

lambda_func = lambda x: True if x**2 >= 10 else False

print(lambda_func(3))
print(lambda_func(4))
```

```python
#Ejemplo 5

mi_dicc = {"A": 1, "B": 2, "C": 3}

sorted(mi_dicc, key=lambda x: mi_dicc[x]/2, reverse=True)
```
