### Condicional 1 (if)

#### La declaración if evalúa una condición y, si es verdadera, ejecuta el bloque de código que le sigue.

```python
if 'condicion':
    # Bloque de código que se ejecuta si la condición es verdadera
```

```python
## Ejemplos

if 6 > 5:
  print("Es verdadero")

if 15 < 10:
  print("15 es menor que 10")

if len("Carlos") == len("Andres"):
  x = 6
  print(x)
```

```python
else:
    # Bloque de código que se ejecuta si todas las condiciones anteriores son falsas
```

```python
elif otra_condicion:
    # Bloque de código que se ejecuta si la otra_condicion es verdadera
```

# Loops

## Bucle For

#### El bucle for en Python se utiliza para iterar sobre una secuencia (que puede ser una lista, una tupla, un diccionario, un conjunto o una cadena).

#### Ejemplos

```python
#1. Lista

lista = [1,2,3,4,5]
for numero in lista:
    print(numero)
```

```python
#2. Diccionario

persona = {
    "nombre": "Juan",
    "edad": 22,
    "ciudad": "San Jose",
    "email": "juan@example.com"
}

for clave, valor in persona.items():
    print(f"{clave}: {valor}")
```

```python
#3. Rango

for i in range(5): # Esto va desde 0 hasta 4
    print(i)
```

```python
#4. For con if y else

numeros = [1, 2, 3, 4, 5, 6]

for numero in numeros:
    if numero % 3 == 0: ## Residuo
        print(f"{numero} es divisible entre 3")
    else:
        print(f"{numero} no es divisible entre 3")
```

```python
#5. For con Enumerate

nombres = ['Juan', 'María', 'Pedro']
for indice, nombre in enumerate(nombres):
    print(indice+1, nombre)
```

```python
#6. For con count
count = 1

nombres = ['Juan', 'María', 'Pedro']
for nombre in nombres:
    print(f'{count} {nombre}')
    count += 1

print(count)
```

## Bucle While

#### El bucle While en Python se utiliza para iterar sobre una secuencia siempre y cuando una la condicion sea verdadera.

```python
#1. Numero

suma = 0
numero = 1
limite = 50

while suma <= limite:
    suma += numero
    numero += 1
    #print(suma)

print(f"La suma de los primeros {numero - 1} números es mayor que {limite} ya que el resultado es {suma}")
```

```python
#2. While con lista

lista = [1, 2, 3, 4, 5, 6, 7, 8]
indice = 0

while indice < len(lista):  ## len = 8
    x = lista[indice]
    if x >= 4:
        print(x, "Es mayor o igual a 4")
    indice += 1
```

```python
## While iterator eterno

count = 0
max = 100

while count < max:
  print(count)
```

```python
#3. While con diccionario

diccionario = {"Pedro":25, "Alex":18, "Luis":19}
indice = 0

while indice < len(diccionario):
    for key,val in diccionario.items():
      if val >= 18:
          print(key, "es mayor de edad")
      indice += 1
```

```python
#4. While con texto

contraseña_correcta = "curso_python_2023"
entrada = ""
count = 0
while entrada != contraseña_correcta:
    entrada = input("Por favor, introduzca su contraseña: ")
    count += 1

    if entrada == contraseña_correcta:

      print("¡Contraseña correcta!")

    if count >= 4:
      print("La contraseña esta incorrecta")
      break
```

## Control de flujo (break, continue, pass)

#### Los controles de flujo se emplean en el momento donde se desea cortar/continuar u omitir algun dato durante alguna iteración o bien durante algún código que tenga alguna condición especial que se desee considerar

```python
#1. Ejemplo con código

for i in range(5):
    if i == 2:
        pass
    print(i)
```

```python
#2. Ejemplo con break

for i in range(5):
  if i == 2:
    break
  print(i)
```

```python
#3. Break en una matriz

matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

valor = 6
found = False

for row in matrix:
  for item in row:
        if item == valor:
            found = True
            print(row)
            break
        else:
            continue
```

```python
#4. Procesar y filtrar

data = [1, 'two', None, 3, 'four', 'five', 'six', 7, 8]
sum_strings = 0

numeros = []
for item in data:
    if isinstance(item, str): ## Si el valor es un string
        sum_strings += 1
        if sum_strings >= 3:
            print("Demasiadas cadenas consecutivas. Abortando.")
            break
        continue
    elif item is None:
        pass  # Ignorar los valores None, pero reseteamos el conteo de cadenas
        sum_strings = 0
    else:
        print(f"Procesando el número {item}")
        numeros.append(item)
        sum_strings = 0

print(numeros)
```

```python
## Iterar sobre el dataframe, extraer los numeros y calcular el promedio

## Crear un for que va a iterar sobre las filas de mi dataframe

lista_datos = []
sum = 0

for col in dataframe.columns:
  for x, row in dataframe.iterrows():
    x = row[col]
    if isinstance(x, float) or isinstance(x, int):
      lista_datos.append(x)
      sum += x
    else:
      continue


  promedio = (sum / len(lista_datos))

print(sum)
print(promedio)
print(len(lista_datos))
```
