```python
## Librerias

import matplotlib.pyplot as plt
import numpy as np
from scipy.misc import derivative
import pandas as pd
```

## Ejemplos

```python
## Gráfico de linea (funcion_seno)

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, label='Función seno')
plt.title('Gráfico de línea')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.legend()
plt.show()
```

```python
## Grafico de lineas ejemplo 2 (varias curvas)

x = np.linspace(0, 10, 100)
sin = np.sin(x)
cos_aprox = np.gradient(sin, x)
cos = np.cos(x)

fig, ax = plt.subplots()
ax.plot(x, sin, label='sin(x)')
ax.plot(x, cos_aprox, label='cos_approx(x)')
ax.plot(x, cos, label='cos(x)')
ax.legend()
ax.set_title('Funciones sin(x) y cos(x)')
ax.set_xlabel('x')
ax.set_ylabel('y')
```

```python
## Grafico de lineas con x y y

x = [2,4,6,8,10,12,14,16]
y = [5,9,14,20,25,29,27,30]

plt.plot(x, y, label='Relacion X y Y')
plt.title('Gráfico de línea')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(x, label='Valores en X')
ax.plot(y, label='Valores en Y')
ax.legend()
ax.set_title('Comparacion de lineas')
ax.set_xlabel('x')
ax.set_ylabel('y')
```

```python
## Grafico de Dispersion

x = [2,4,6,8,10,12,14,16]
y = [5,9,14,20,25,29,27,30]

plt.scatter(x, y, color='blue', marker='v')
plt.title('Gráfico de dispersión')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.show()
```

```python
# Fixing random state for reproducibility
np.random.seed(19680801)


def randrange(n, vmin, vmax):
    """
    Helper function to make an array of random numbers having shape (n, )
    with each number distributed Uniform(vmin, vmax).
    """
    return (vmax - vmin)*np.random.rand(n) + vmin

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

n = 100

# For each set of style and range settings, plot n random points in the box
# defined by x in [23, 32], y in [0, 100], z in [zlow, zhigh].
for m, zlow, zhigh in [('o', -50, -25), ('^', -30, -5)]:
    xs = randrange(n, 23, 32)
    ys = randrange(n, 0, 100)
    zs = randrange(n, zlow, zhigh)
    ax.scatter(xs, ys, zs, marker=m)

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```

```python
## Grafico de barras

labels = ['Luis', 'Alex', 'Maria', 'Daniela']
values = [12, 38, 30, 40]

plt.bar(labels, values, color=['red', 'green', 'blue', 'black'])
plt.title('Gráfico de barras')
plt.xlabel('Etiquetas')
plt.ylabel('Valores')
plt.show()
```

```python
## Grafico de Histogramas

for i in range (1000,100000,1000):
  data = np.random.randn(i)

  plt.hist(data, bins=30, color='blue', alpha=0.3)
  plt.title('Histograma con {} datos'.format(i))
  plt.xlabel('Valores')
  plt.ylabel('Frecuencia')
  plt.show()
```

```python
## Grafico circular

def circular_plot(labels_, sizes_, colors_,ratio):
  plt.pie(sizes_, labels=labels_, colors=colors_,
  autopct='%1.1f%%', shadow=True, startangle=ratio)

  plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
  plt.title('Gráfico circular')
  plt.show()

labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
colors = ['red', 'blue', 'yellow', 'green']

#circular_plot(labels, sizes, colors)

labels2 = ['Masculino','Femenino']
sizes2 = [50,50]
colors2 = ['blue', 'pink']

circular_plot(labels2,sizes2,colors2,120)
```

```python
## Saber la distribucion en la columna de genger_binary

gender_distribution = list(df['gender_binary'].value_counts())
gender_distribution

def circular_plot(labels_, sizes_, colors_,ratio):

  def intern_func(pct,sizes_all):
    absolute = int(pct/100 * np.sum(sizes_all))
    return "{:.1f}%\n{:d}".format(pct, absolute)

  plt.pie(sizes_, labels=labels_, colors=colors_,
  autopct=lambda pct: intern_func(pct,sizes_),  startangle=ratio)

  plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
  plt.title('Gráfico circular')
  plt.show()

labels3 = ['Masculino','Femenino']
circular_plot(labels3,gender_distribution,colors2,120)
```

```python
## Group by por genero mujeres y desplegar las distribuciones por color

group_mujeres = df[df['gender_binary'] == 'Female']


distribucion_mujeres = dict(group_mujeres['color'].value_counts())


colors = []

for key, value in distribucion_mujeres.items():
  colors.append(key)

print(colors)
colors[4] = 'Gray'
colors[14] = 'Black'
colors[-2] = 'Blue'


#distribucion_mujeres

plt.figure(figsize=(15,10))

for key, value, color in zip(distribucion_mujeres.keys(), distribucion_mujeres.values(), colors):

  plt.bar(key, value, color=color)

plt.title('Gráfico de barras')
plt.xlabel('Colores')
plt.ylabel('Valores')
plt.show()
```
